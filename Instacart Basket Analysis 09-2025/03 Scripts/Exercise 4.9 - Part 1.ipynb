## Data Wrangling

```python
# Most of the headers for fields do not follow typical naming structure, contain typos, or are unclear so we'll do some renaming first

df_cust.rename(columns = {'First Name' : 'first_name', 'Surnam' : 'last_name', 'Gender' :'gender', 'STATE' : 'state', 'Age': 'age',
                          'n_dependants' : 'num_of_dependents', 'fam_status' : 'martial_status'}
                          , inplace = True)
```

```python
df_cust.columns
```

```python
# Next, we'll drop the fields first_name and last_name since this is personally identifiable information (PII)
# so it's best remove this information from the dataframe for privacy concerns

df_cust = df_cust.drop(columns = ['first_name', 'last_name'])
```

```python
df_cust.shape
```

## Data Integrity and Consistency Checks

### Null Check

```python
# Based on our code here, there are no null values associated with any of the fields

df_cust.isnull().sum()
```

### Duplicate Checking

```python
# We'll set up a subset to see if duplicates are present in the dataframe

df_dups = df_cust[df_cust.duplicated()]
```

```python
# Our result shows that there are no exact match duplicates so can can proceed with cleaning

df_dups
```

### Checking for Mixed Types

```python
# Despite the warnings, this shows us no mixed-type data fields are present

for col in df_cust.columns.tolist():
    weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_cust[weird]) > 9:
        print(col)
```
